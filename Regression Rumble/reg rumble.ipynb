```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

from sklearn.svm import SVC
from sklearn import svm
```

```python
data = pd.read_csv('data.csv')
test = pd.read_csv('test.csv')
sam_sub = pd.read_csv('sample_submission.csv')
```

```python
def plot_histograms(dataframe):
    # Loop through each column in the DataFrame
    for column in dataframe.columns:
        # Create a histogram for the current column
        plt.figure()
        sns.distplot(dataframe[column], bins=10)  # Adjust the number of bins as needed
        plt.title(f"Histogram of {column}")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.show()
```

```python
plot_histograms(data)
```

```python
feat = ['x0', 'x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9']
target = ['y']
```

```python
X = data[feat]
y = data[target]
```

```python
#Test train split
# Split dataset into training and test sets (80/20 split)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
train = X_train.copy()
train['y'] = y_train

val = X_val.copy()
val['y'] = y_val
```

```python
#ALL SCORES
def scores(y_true, y_pred):
    from sklearn.metrics import mean_absolute_error
    from sklearn.metrics import mean_squared_error
    from sklearn.metrics import r2_score
    
    r2 = r2_score(y_true, y_pred)

    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)

    mae = mean_absolute_error(y_true, y_pred)
    
    print(f" R2 Score is {r2}")
    print (f" Mean Sq Erro is {mse}")
    print (f" Root MSE is {rmse}")
    print (f" Mean Abs Error is {mae}")
```

```python
#1 LINEAR MODEL
lin_reg = LinearRegression(fit_intercept=True, normalize = False)
lin_reg.fit(train[feat], train[target])
```

```python
# use trained model to make predictions on training data
train['lin_pred'] = lin_reg.predict(train[feat])
scores(train[target],train['lin_pred'])
```

```python
# use trained model to make predictions on VAL data
val['lin_pred'] = lin_reg.predict(val[feat])
scores(val[target],val['lin_pred'])
```

```python
test['y'] = lin_reg.predict(test[feat])
my_sub = test[['id','y']]
ss = my_sub.to_csv('Lin_reg.csv')
```

```python
#2 POLYNOMIAL 
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(train[feat])

# create LinearRegression object and fit on transformed training data
poly_reg = LinearRegression(fit_intercept=True, normalize = False)
poly_reg.fit(X_poly, train[target])

# use trained model to make predictions on training data
train['poly_pred'] = poly_reg.predict(X_poly)
```

```python
scores(train[target],train['poly_pred'])
```

```python
X_poly_val = poly.transform(val[feat])

# use trained model to make predictions on VAL data
val['poly_pred'] = poly_reg.predict(X_poly_val)
scores(val[target],val['poly_pred'])
```

```python
X_poly_test = poly.transform(test[feat])
test['poly_pred'] = poly_reg.predict(X_poly_test)
```

```python
ply_sub = test[['id','poly_pred']]
ply_sub = ply_sub.to_csv('ply_sub.csv')
```
